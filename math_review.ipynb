```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### معادلة الخط المستقيم
### Line Equation

```python
#you can change b and a and run the code to see the effect
b = 1 #intercept
a = 1.5 #slope
points = np.linspace(0,10,10)
line = b + a*points
plt.plot(line)
plt.ylim(0, 10)
```

## كثيرات الحدود (حدين)
## Polynomial (of degree 2)

```python
#you can change a0, a1, and a2 and run the code to see the effect
a0 = 1 
a1 = -0.2 
a2 = 0.2
points = np.linspace(-10,10,100)
curve = a0 + a1*points + a2*points**2
plt.plot(curve)
plt.ylim(0, 10)
```

## كثيرات الحدود (ثلاثة حدود)
## Polynomial (of degree 3)

```python
#you can change a0-a3 and run the code to see the effect
a0 = 1
a1 = -.2 
a2 = 2
a3 = -0.5
points = np.linspace(-10,10,100)
curve = a0 + a1*points + a2*points**2 + a3*points**3
plt.plot(curve)
plt.ylim(-10, 10)
```

## كثيرات الحدود (أربعة حدود)
## Polynomial (of degree 4)

```python
#you can change a0-a4 and run the code to see the effect
a0 = 1
a1 = -.2 
a2 = 2
a3 = -0.4
a4 = -0.3
points = np.linspace(-10,10,100)
curve = a0 + a1*points + a2*points**2 + a3*points**3 + a4*points**4
plt.plot(curve)
plt.ylim(-10, 10)
```

## المعادلات الموجية، يمكن عن طريقها تمثيل التغير المتكرر عبر الزمن
## Sine and Cosine Wave Functions

```python
#sine and cosine waves, for periodic occurances
points = np.linspace(0,10,100)
wave_sin = np.sin(points)
wave_cos = np.cos(points)
plt.plot(wave_sin)
plt.plot(wave_cos, 'r--')
plt.ylim(-1.5, 1.5)
```

## من المهم معرفة حساب المسافة بين النقاط، هذا الكود يوضح طريقة بسيطة ومشهورة لذلك
## مثالين لحساب المسافة بين نقطتين في بعد ثنائي وأبعاد كبيرة

```python
#calculate distances between two points using Euclidean distance: sum(sqrt((point1-point2)**2)
point1 = np.array([2, 4])
point2 = np.array([8, 2])

#visualize the two points
plt.scatter(point1, point2)

#Euclidean distance
distance = sum(np.sqrt((point1-point2)**2))

print('Euclidean distance between two points (2D) = {}'.format(distance))

#can be generalize to points of any dimension
point1 = np.array([12, 32, 11, 34, 56])
point2 = np.array([1, 12, 34, 23, 4])

distance = sum(np.sqrt((point1-point2)**2))
print('Euclidean distance between two points = {}'.format(distance))
```
